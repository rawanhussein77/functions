```python
import dash
from dash import dcc, html
import pandas as pd
import plotly.graph_objects as go

data = pd.read_csv(r"/country_wise_latest (1).csv")


missing_values = data.isnull().astype(int)

fig = go.Figure(
    data=go.Heatmap(
        z=missing_values.values,
        x=missing_values.columns,
        y=[f"Row {i}" for i in range(len(missing_values))],
        colorscale="Blues",
        showscale=True
    )
)


fig.update_layout(
    title="Heatmap of Missing Values",
    xaxis_title="Columns",
    yaxis_title="Rows",
    template="plotly"
)


app = dash.Dash(__name__)


app.layout = html.Div([
    html.H1("COVID-19 Missing Values Heatmap", style={"textAlign": "center"}),


    dcc.Graph(figure=fig, id="heatmap-graph")
])

if __name__ == "__main__":
    app.run_server(debug=True)
```

```python
import plotly.express as px
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import pandas as pd


countries = ['Italy', 'Yemen', 'Albania', 'Andorra', 'Zimbabwe', 'Algeria']
filtered_data = data[data['Country/Region'].isin(countries)]

variables = ['Confirmed', 'Deaths', 'Recovered', 'Active']

app = dash.Dash(__name__)


app.layout = html.Div([
    html.H1("COVID-19 Data Distribution"),


    dcc.Dropdown(
        id='country-dropdown',
        options=[{'label': country, 'value': country} for country in countries],
        value=countries,
        multi=True,
        style={'width': '50%'}
    ),


    dcc.Graph(id='confirmed-graph'),
    dcc.Graph(id='deaths-graph'),
    dcc.Graph(id='recovered-graph'),
    dcc.Graph(id='active-graph')
])

@app.callback(
    [Output('confirmed-graph', 'figure'),
     Output('deaths-graph', 'figure'),
     Output('recovered-graph', 'figure'),
     Output('active-graph', 'figure')],
    [Input('country-dropdown', 'value')]
)
def update_graphs(selected_countries):

    filtered_data = data[data['Country/Region'].isin(selected_countries)]


    fig_confirmed = px.histogram(filtered_data, x='Confirmed', color='Country/Region', nbins=10, title='Distribution of Confirmed Cases')
    fig_deaths = px.histogram(filtered_data, x='Deaths', color='Country/Region', nbins=10, title='Distribution of Deaths')
    fig_recovered = px.histogram(filtered_data, x='Recovered', color='Country/Region', nbins=10, title='Distribution of Recovered Cases')
    fig_active = px.histogram(filtered_data, x='Active', color='Country/Region', nbins=10, title='Distribution of Active Cases')

    return fig_confirmed, fig_deaths, fig_recovered, fig_active

if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import pandas as pd

import plotly.graph_objects as go

app = dash.Dash(__name__)

numerical_data = data.select_dtypes(include=['number'])
numerical_data = numerical_data.drop(columns=['WHO Region'], errors='ignore')

corr_matrix = numerical_data.corr()

fig = go.Figure(data=go.Heatmap(
    z=corr_matrix.values,
    x=corr_matrix.columns,
    y=corr_matrix.columns,
    colorscale=[[0, 'blue'], [1, 'red']],
    zmin=-1, zmax=1,
    colorbar=dict(title='Correlation Coefficient')
))


app.layout = html.Div([
    html.H1('Correlation Matrix of COVID-19 Data', style={'text-align': 'center'}),
    dcc.Graph(
        id='correlation-heatmap',
        figure=fig
    )
])


if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.express as px
import pandas as pd
import plotly.graph_objects as go
from dash.dependencies import Input, Output


country_deaths = data.groupby('Country/Region')['Deaths'].sum().reset_index()
country_deaths = country_deaths.sort_values(by=['Deaths'], ascending=False)
top_10_countries = country_deaths.head(10)

fig = px.bar(top_10_countries, x='Country/Region', y='Deaths',
             title='Total Deaths by Country (Top 10)',
             labels={'Deaths': 'Total Deaths', 'Country/Region': 'Country/Region'})

app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("COVID-19 Deaths Dashboard", style={'text-align': 'center'}),
    dcc.Graph(id='death-bar-chart', figure=fig),
])

if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.express as px
import pandas as pd
from dash.dependencies import Input, Output

country_recoveries = data.groupby('Country/Region')['Recovered'].sum().reset_index()


country_recoveries = country_recoveries.sort_values(by=['Recovered'], ascending=False)
top_10_countries = country_recoveries.head(10)

fig = px.bar(top_10_countries, x='Country/Region', y='Recovered',
             title='Total Recoveries by Country (Top 10)',
             labels={'Recovered': 'Total Recoveries', 'Country/Region': 'Country/Region'})


app = dash.Dash(__name__)


app.layout = html.Div([
    html.H1("COVID-19 Recoveries Dashboard", style={'text-align': 'center'}),


    dcc.Dropdown(
        id='num-countries',
        options=[{'label': str(i), 'value': i} for i in range(1, 21)],
        value=10,
        style={'width': '40%', 'margin': 'auto'}
    ),


    dcc.Graph(id='recovery-bar-chart', figure=fig),
])

@app.callback(
    Output('recovery-bar-chart', 'figure'),
    [Input('num-countries', 'value')]
)
def update_graph(num_countries):

    top_countries = country_recoveries.head(num_countries)


    fig = px.bar(top_countries, x='Country/Region', y='Recovered',
                 title=f'Total Recoveries by Country (Top {num_countries})',
                 labels={'Recovered': 'Total Recoveries', 'Country/Region': 'Country/Region'})

    return fig


if __name__ == '__main__':
    app.run_server(debug=True)
```
